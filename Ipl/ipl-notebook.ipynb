### **1. Top batsman each season**

```python
df = deliveries.merge(matches[['id', 'season']], left_on='match_id', right_on='id')
season_batsman_runs = df.groupby(['season', 'batter'])['batsman_runs'].sum().reset_index()

top_batsman_per_season = season_batsman_runs.sort_values(['season', 'batsman_runs'], ascending=[True, False]).groupby('season').head(10)
top_batsman_per_season.to_csv('top_batsman_per_season.csv')
```

### **3.Batting first vs Chasing win %**

```python
matches.head()
```

```python
def get_batting_first(row):
  if row['toss_decision'] == 'bat':
    return row['toss_winner']
  else:
    return row['team2'] if row['toss_winner'] == row['team1'] else row['team1']
```

```python

# Step 2: Apply to the matches DataFrame
matches['batting_first'] = matches.apply(get_batting_first, axis=1)

# Step 3: Define chasing team
matches['chasing_team'] = matches.apply(lambda row: row['team1'] if row['batting_first'] != row['team1'] else row['team2'], axis=1)

# Step 4: Identify whether the winner was batting or chasing
matches['win_type'] = matches.apply(
    lambda row: 'Batting First' if row['winner'] == row['batting_first']
    else ('Chasing' if row['winner'] == row['chasing_team'] else 'No Result'), axis=1
)

# Step 5: Count win percentages
win_type_counts = matches['win_type'].value_counts(normalize=True) * 100
win_type_df = win_type_counts.reset_index()
win_type_df.columns = ['type', 'win_percent']

# Step 6: Save for Tableau
win_type_df.to_csv('batting_vs_chasing_win_percent.csv', index=False)
win_type_df


```

#### Toss win vs Match win


```python
matches['toss_match_result'] = matches.apply(lambda row: 'Yes' if row['toss_winner'] == row['winner'] else 'No', axis=1)
toss_match = matches['toss_match_result'].value_counts(normalize=True).reset_index()

toss_match.columns = ['toss_match_result', 'Win percent']
toss_match.to_csv('toss_vs_match_wins.csv', index=False)
```
